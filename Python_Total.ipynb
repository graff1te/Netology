# Задание 7.
Проверить гипотезу, что сотрудники с высоким окладомпроводят на работе больше времени, чем сотрудники с низкимокладом

```python
low_salary = df[df['salary'] == 'low'] [['average_montly_hours']]
low_salary
```

```python
high_salary = df[df['salary'] == 'high'] [['average_montly_hours']]
high_salary
```

```python
# Визуально анализируем нормальность рапределения
plt.hist(low_salary, label='low_salary')
plt.hist(high_salary, label='high_salary')
plt.legend()
plt.show()
```

H0 = сотрудники с высоким и низким окладом проводят на работе одинаковое время.

```python
t, p_value = stats.ttest_ind(high_salary, low_salary)  # Критерий Стьюдента (t-test)
t, p_value
```

p-value = 46%  
Не можем отклонить H0 на достаточно высоком уровне значимости (до 46%)

# Задание 9.

- Разделить данные на тестовую и обучающую выборки  
- Построить модель LDA, предсказывающую уволился лисотрудник на основе имеющихся факторов (кроме department иsalary)  
- Оценить качество модели на тестовой выборки

```python
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
X = df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
y = list(df['left'])
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=555)
```

```python
model = LinearDiscriminantAnalysis()
model.fit(X_train, y_train)  # метод обучается на данных и подбирает оптимальные коэффициенты
```

```python
# метод возвращает значение коэффициента детерминации

model.score(X_test, y_test)
```

```python
# расчет точности - отношение верных прогнозов к общему количеству позиций
# является хорошей метрикой для сбалансированных классов 
# (наблюдений в категориях примерно одинаковое количество)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, model.predict(X_test))
```

В нашем случае ~76% изменчивости зависимого выходного атрибута можно объяснить моделью,  
в то время как остальные 24% изменчивости все еще не учтены.

```python
# метод возвращает прогнозные значения на тестовых данных модели

y_pred = model.predict(X_test)
y_pred
```

```python
# смотрим разницу факта и прогноза

result = pd.DataFrame([y_test, model.predict(X_test)]).T
result
```
